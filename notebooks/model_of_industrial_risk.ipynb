```python
import random
def syntetic_aug_dataset(count=100, data_mother=None):
    dic_new_data = {}
    for col in data_mother.columns:
        dic_new_data[col] = random.choices(data_mother[col].tolist(), k=count)
    return  pd.concat([pd.DataFrame(dic_new_data), data_mother], axis=0)
```

```python
augment_data = syntetic_aug_dataset(count=10000, data_mother=raw_data_set)
augment_data.shape
```
